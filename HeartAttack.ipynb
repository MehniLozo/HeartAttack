```python
import  pandas as pd
# data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataf=pd.read_csv('heart-1.csv')
print("dataf",type(dataf))
head=dataf.head()
print("head",type(head))
print(head)
print(dataf.info())
print(dataf.describe())
```

```python
ax=sns.countplot(x='sex', data=dataf)
ax.set(xlabel='1=Male, 0=Female',ylabel='Counts',title='Gender distribution')
plt.show()
```

```python
#distrubition exang: exercise induced angina (1 = yes; 0 = no)
ax=sns.countplot(x='exng', data=dataf)
ax.set(xlabel='1=yes, 0=no',ylabel='Counts',title='Exercise induced angina distribution')
plt.show()
```

```python
#Distribution of Chest Pain Type
ax=sns.countplot(x='cp', data=dataf)
ax.set(xlabel='1=typical, 2=atypical, 3=non-anginal, 4=asymptomatic',ylabel='Counts',title='distribution of chest pain type')
plt.show()
```

```python
#Gender distribution of people having Exercise Induced Asthama
ax=sns.countplot(x='sex',hue='output',data=dataf)
plt.show()
```

```python
ax=sns.countplot(x='output',hue='sex',data=dataf)
plt.show()
```
